```python
%matplotlib inline
import numpy as np
df1 = df[['Period','Data_value']]
bins =  np.arange(0,5000,1000)
groups = df1.groupby(pd.cut(df1['Period'],bins)).mean()
print(groups.head())
groups['Period'].plot.line()
```

```python
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()

X = df[['Period', 'MAGNTUDE']]
y = df['Data_value']

X[['Period', 'MAGNTUDE']] = scale.fit_transform(X[['Period', 'MAGNTUDE']])

print (X)

est = sm.OLS(y, X).fit()

est.summary()
```

```python
y.groupby(df.STATUS).mean()
```

```python
scaled = scale.transform([[3000, 6]])
print(scaled)
predicted = est.predict(scaled[0])
print(predicted)
```
